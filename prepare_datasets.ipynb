```python
import pandas as pd

woman_abc_df = pd.read_csv("women_at_ABC_26_11_2020.csv")
members_onmc_df = pd.read_csv("people_at_onmc_in_2007_26_11_2020.csv")
```

```python
woman_abc_onc_df = pd.merge(woman_abc_df, members_onmc_df, on=['name'], how='inner')
```

```python
woman_abc_onc_df
```

```python
import wikipedia
wikipedia.set_lang("pt")

wiki_dict = {}
for i, row in woman_abc_onc_df.iterrows():

    try:
        page = wikipedia.page(row["name"])
        wiki = page.url
    except:
        wiki = ""

    wiki_dict[row["name"]] = wiki
```

```python
woman_abc_onc_df["wikipedia"] = woman_abc_onc_df["name"].map(wiki_dict)

woman_abc_onc_df
```

```python
woman_abc_onc_df.to_csv("women_at_abc_and_onmc_df_26_11_2020.csv")
```
