```python
import pandas as pd

man_abc_onc_df = pd.read_csv("man_abc_onc_df_01_12_2020.csv")
```

```python
import wikipedia
wikipedia.set_lang("pt")

wiki_dict = {}
for i, row in man_abc_onc_df.iterrows():
    print(i)

    try:
        page = wikipedia.page(row["name"])
        wiki = page.url
    except:
        wiki = ""

    wiki_dict[row["name"]] = wiki
```

```python
man_abc_onc_df["wikipedia"] = man_abc_onc_df["name"].map(wiki_dict)

man_abc_onc_df
```

```python
man_abc_onc_df.to_csv("men_at_abc_and_onmc_and_wiki_df_01_12_2020.csv")
```
